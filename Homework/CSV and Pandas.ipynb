```python
import pandas as pd
```

```python
#1 Reading the CSV File:
df = pd.read_csv('/Users/diegopretelt/Desktop/DSS/HomeWork/CSV and Pandas/acc_players-2324F.csv', skiprows = 1)
df.head() #removed header row as it provided confusing names
```

```python
#2 Basic Analysis:
total_points = df['PTS'].sum()
print(total_points)

most_minutes = df.loc[df['MP'].idxmax()]
print(most_minutes['Player'])

top5_rebounds = df[['Player','TRB']].sort_values('TRB',ascending = False).head(5)
print(top5_rebounds)
```

```python
# 3 Player Filtering:
small_df = df[df['MP'] > 500]
small_df.head()

most_asst = df.loc[df['AST'].idxmax()]
print(most_asst['Player'])

top3_asst = df[['Player','AST']].sort_values('AST',ascending = False).head(3)
print(top3_asst)

top3_shtblk = df[['Player','BLK']].sort_values('BLK',ascending = False).head(3)
print(top3_shtblk)
```

```python
# 4 School-Based Analysis:
school_pts = df.groupby('School')['PTS'].sum()
print(school_pts)

school_ast = df.groupby('School')['AST'].sum()
print(school_ast)

top3_schools = school_pts.sort_values(ascending=False).head(3)
print(top3_schools)
```
